#### [日盛證券_台積電(2330)資產負債表](http://jsjustweb.jihsun.com.tw/z/zc/zcp/zcpa/zcpa0.djhtm?b=Y&a=2330)
>table 標籤中, 並非正規的 tr, td. 而是 span, pd.read_html 讀取時無法得到我們想要的資料, 如下所示:

```python
import pandas as pd
url = 'http://jsjustweb.jihsun.com.tw/z/zc/zcp/zcpa/zcpa0.djhtm?b=Y&a=2330'
df = pd.read_html(url)
print(len(df))
```

```python
df[0]
```

```python
df[1]
```

```python
df[2]
```

```python
import requests
from bs4 import BeautifulSoup
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
rows = soup.select('#oMainTable .table-row')
print(len(rows))
# ★★★ 若將 'html.parser' 改為 'lxml' → 可能會造成取到的資料筆數變少 → 宜注意
```

```python
rows[0]
```

```python
columns = list()
for x in rows[0].select(".table-cell"):
    columns.append(x.text)
columns
```

```python
trs = list()
for x in rows[2:]: # 第二列不取
    tds = list()
    for y in x.select(".table-cell"):
        tds.append(y.text)
    trs.append(tds)
```

```python
trs[0] # \u3000 → 全形的空白
```

```python
df = pd.DataFrame(trs)
df.columns = columns
df.head(10)
```
