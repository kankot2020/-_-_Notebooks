## Series 一維資料

```
se = pd.Series(串列)
```

範例資料:
```
stocks = ['聯電', '台積電', '聯發科', '大力光', '鴻海']
price = [42, 510, 694, 2115, 109]
```

````python
import pandas as pd

price = [42, 510, 694, 2115, 109] # 一維資料 Series
se = pd.Series(price)
print(se, end = "\n\n")
print("求值:", se.values) # 求值
print("索引:", se.index) # 求索引
print("索引轉List:", list(se.index))
````

````python
# 自訂索引
stocks = ['聯電', '台積電', '聯發科', '大立光', '鴻海']
price = [42, 510, 694, 2115, 109]
se2 = pd.Series(price, index=stocks)
se2
````

````python
# 資料資訊 shape, describe()
se2.shape
se2.describe()
````

````python
# 統計：加總、最小值、最大值、平均值、中位數
print(f"最小值:{se2.min()}, 最大值:{se2.max()}")
print(f"加總:{se2.sum()}, 平均值:{se2.mean()}, 中位數:{se2.median()}")
````

````python
se2
````

````python
# Series 取值
# 用索引值(排的順序, 由零開始)
se2[1]  # 台積電
se2[3]  # 大立光

# iloc(用索引值), loc(用索引名稱)
se2.iloc[1]
se2.loc['大立光']

# 一次取多個值
se2.iloc[[1, 3, 4]]
se2.loc[["台積電", "大立光", "鴻海"]]

# 一次取多個連續值, iloc 用 range() 的觀念
# se2.iloc[1:4]               # ★★ 包含頭, 不含尾 ★★
# se2.loc["台積電":"大立光"]   # ★★ 包含頭, 包含尾 ★★
````

## DataFrame 二維資料

```
df = pd.DataFrame(字典)
```

範例資料:

```
{
  "代號":['2303', '2330', '2454', '3034', '2317'], "公司":['聯電', '台積電', '聯發科', '聯詠', '鴻海'],
  "開盤價":[40,507,702,256,108],  "最高價":[41,511,706,262,109],
  "最低價":[40,504,690,253,107],  "收盤價":[41,510,694,260,109] 
}
```

````python
# 定義 DataFrame 資料
import pandas as pd
df = pd.DataFrame({
  "代號": ['2303', '2330', '2454', '3034', '2317'],
  "公司": ['聯電', '台積電', '聯發科', '聯詠', '鴻海'],
  "開盤價": [40, 507, 702, 256, 108],
  "最高價": [41, 511, 706, 262, 109],
  "最低價": [40, 504, 690, 253, 107],
  "收盤價": [41, 510, 694, 260, 109] 
})
df
````

````python
# 顯示 index, columns, 取得所有值
print("◆index:\n", df.index)
print("◆columns:\n", df.columns)
print("◆values:\n", df.values)
````

````python
# 資料資訊 shape, describe(), info()
print("◆shape:\n", df.shape)
print("◆info:")
df.info()
df.describe()
````

````python
df
````

````python
# 取得列資料: 1.索引 2.iloc(索引值) 3.loc(索引名)
df.values[1] # ★ 不可以輸入為 df[1] ★
df.iloc[1]
df.loc[1]

# 取得欄資料(單欄, 多欄)
df['公司']
df['收盤價']
df[['公司', '收盤價']]
df[['代號', '開盤價', '收盤價']]

# 取得指定欄列(儲存格)的值: 1.索引 2.iloc(索引值) 3.loc(索引名)
df.values[1, 2]
df.iloc[1, 2]
df.loc[1, '開盤價']
````

````python
# 進階取值: 多列、多欄、連續列、連續欄
# 任務: 1.取得前三筆資料 2.取得前三筆資料的公司名
# ====================================================================
# 方法一: df.values[index, column] 二維陣列取值，回傳值為array
df.values[0:3]      # ◆ 資料格式: 二維陣列 
df.values[0:3, 1]   # ◆ 資料格式: 一維陣列
# ====================================================================
# 方法二: df.iloc[index, column]
# df.iloc[0:3]      # ◆ 資料格式: DataFrame 
# df.iloc[0:3, 1]   # ◆ 資料格式: Series
# ====================================================================
# 方法三: df.loc[index_name, column_name]
df.loc[0:2]
df.loc[0:2, '公司']

# ★ iloc, loc 取得的資料便於顯示, 格式漂亮。
# ★ 若要將資料給其它程式, 則以 df.values 得到的一維/二維陣列為佳。
````

## 資料連接與合併

* concat()：水平或垂直連接
* merge()：合併

````python
# 統一發票中獎號碼
# 本期 https://invoice.etax.nat.gov.tw/index.html
# 上期 https://invoice.etax.nat.gov.tw/lastNumber.html

import pandas as pd
url1 = 'https://invoice.etax.nat.gov.tw/index.html'
url2 = 'https://invoice.etax.nat.gov.tw/lastNumber.html'
df1 = pd.read_html(url1)[0]
df2 = pd.read_html(url2)[0]
````

````python
df1
````

````python
# concat() "預設為垂直串接, axis=0, ignore_index=True 重編索引"
df_ver = pd.concat([df1, df2], ignore_index=True)
df_ver
````

````python
# concat() 水平串接 → axis=1
df_hor = pd.concat([df1, df2], axis=1)
df_hor
````

````python
# 資料整理: 標題, 去除最後一列
df1.columns = ['獎別', '0506月中獎號碼']
df2.columns = ['獎別', '0304月中獎號碼']
df1.drop(df1.index[-1], axis=0, inplace=True)
df2.drop(df2.index[-1], axis=0, inplace=True)
````

````python
# merge()相同欄位合併
df5 = pd.merge(df1, df2)
df5
````

````python
# 爬取今年一到五月所有的資料 v2
# 台灣證券交易所之限制: 每五秒鍾可以有三次的請求 Requests
import pandas as pd
import time

df_all = pd.DataFrame() # 新增空的DataFrame

for i in range(1, 6):
    url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&' \
         f'date=20220{i}01&stockNo=2330'
    df_temp = pd.read_html(url, header=1)[0]
    df_all = pd.concat([df_all, df_temp], ignore_index=True)
    time.sleep(3)

df_all   
````

# 資料分組 (group by)

資料在指定欄位上進行分組的動作進行統計

````python
# 計算每個縣市的總確診人數
df1.groupby('縣市')['確定病例數'].sum().plot(kind='bar', figsize=(8,4))
````

````python
# 計算每個縣市的確診通報次數
df1.groupby('縣市')['確定病例數'].count().plot(kind='bar', figsize=(8, 4))
````

````python
# 找出每個縣市單週最多確診的人數
df1.groupby('縣市')['確定病例數'].max()
````

````python
# 找出每個縣市單週最少確診的人數
df1.groupby('縣市')['確定病例數'].min()
````

````python
# 計算每個(指定)縣市各鄉鎮的總確診人數
df2 = df1[df1['縣市'] == '新北市']
df2.groupby(['縣市', '鄉鎮'])['確定病例數'].sum().plot(kind='bar', figsize=(8,4))
````

````python
# 經由 groupby() 彙總後資料不是 DataFrame 的型式, 要用 reset_index() 轉回 DataFrame
df2 = df1[df1['縣市'] == '嘉義市']
df2.groupby(['縣市', '鄉鎮'])['確定病例數'].sum().reset_index()
````

````python
# 顯示每個縣市確診人數的各個統計值 agg()，會以 pandas 表格樣式呈現
df1.groupby('縣市')['確定病例數'].agg(['count', 'sum', 'max', 'min'])
````

# 樞紐分析表 (Pivot Table)

在欄及列上同時進行分組的動作

````python
# 檢視 各"縣市" 各"研判年份" 的 "確定病例數" 總合 
pd.pivot_table(df1, values='確定病例數', index='縣市', 
               columns='研判年份', aggfunc='sum', margins=True, margins_name='合計', fill_value=0)
````

# 交叉查詢表 (Cross Table)

用來顯示兩個欄位的值的次數分佈（frequency distribution ），可以用於找出兩個欄位間是否有關聯。

````python
# 檢視 "縣市" "年齡層" 關聯性
pd.crosstab(df1['縣市'], df1['年齡層'])
````

````python
# 檢視 "縣市" "研判年份" 關聯性
pd.crosstab(df1['縣市'], df1['研判年份'])
````

````python
# 檢視 "縣市" "性別" 關聯性
pd.crosstab(df1['縣市'], df1['性別'])

# ​聯合國下屬國際民用航空組織（ICAO）已經更新了有關標凖，規定機場的護照檢查機器的設置中必須增加一個性別選項，X，意思是不詳、不註明、未註明。
````
